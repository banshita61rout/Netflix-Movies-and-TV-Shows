# Step -1 loading data

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
netflix_df=pd.read_csv("netflix_titles.csv")
netflix_df.head(5)
```

# Step -2 Understanding data

```python
netflix_df.shape
```

```python
netflix_df.columns
```

# Step -3 Tansforming correct Col_names 

```python
new_netflix=netflix_df.copy()
```

```python
#Changing Column Names
new_netflix.rename(columns={
    "title":"Title",
    "director":"Director",
    "cast":"Cast",
    "country":"Country",
    "rating":"Rating",
    "release_year":"Release_Year",
    "duration":"Duration",
    "description":"Description",
    "type":"Type",
    "show_id":"Id",
    "date_added": "Date",
    "listed_in": "Genre"
}, inplace=True)

new_netflix.columns
```

# step -4 Checking missing values and replacing them

```python
new_netflix.isnull().sum()
```

```python
#cleaing data - Replacing Missing val

new_netflix["Director"] = new_netflix["Director"].fillna("Unknown")
new_netflix["Cast"] = new_netflix["Cast"].fillna("Unknown")
new_netflix["Country"] = new_netflix["Country"].fillna("Not Specified")
new_netflix["Date"] = new_netflix["Date"].fillna("Not Available")
new_netflix["Rating"] = new_netflix["Rating"].fillna("Not Rated")
new_netflix["Duration"] = new_netflix["Duration"].fillna("Not Available")
```

```python
new_netflix.isnull().sum()
```

## No Null Values detected

# Step -5 Checking duplication

```python
#Duplication check
new_netflix.duplicated().sum()
```

## No Duplication in dataset

# Step -5 Changing date to datetime

```python
# Convert Date column to datetime
new_netflix["Date"] = pd.to_datetime(new_netflix["Date"],errors='coerce')

new_netflix["Year_Added"] = new_netflix["Date"].dt.year
new_netflix["Month_Added"] = new_netflix["Date"].dt.month
new_netflix["Day_Added"] = new_netflix["Date"].dt.day
new_netflix
```

```python
#storing sorted data
new_netflix.to_csv("sorted_netflix_data.csv") 
```

# Step -6 Visualization & Analysis

```python
sns.set_theme()
```

```python
# Content Distribution Movie vs TV Show
sns.countplot(
    data=new_netflix,
    x="Type",
    hue="Type",
    legend=True,
    palette={"Movie": "#FFD166", "TV Show": "lightblue"}
)

plt.title("Content Distribution Movie vs TV Show")
plt.xlabel("Type")
plt.ylabel("")
plt.tight_layout()
```

### Analysis-
The content distribution on Netflix shows a clear dominance of movies over TV shows, with approximately 6,000+ movies compared to around 2,600–2,800 TV shows, meaning nearly 70% of the platform’s content consists of movies while about 30% are TV shows. This indicates that Netflix focuses on building a large and diverse movie library to attract a wide audience quickly, as movies are faster to produce and consume. At the same time, although fewer in number, TV shows play a crucial role in keeping users engaged for longer periods through binge-watching, highlighting a strategic balance between content volume (movies) and user retention (TV shows).

```python
#Growth of Netflix Content Over Time
year_data = new_netflix["Year_Added"].value_counts().sort_index()

sns.lineplot(
    x=year_data.index,
    y=year_data.values,
    marker="o"
)

plt.title("Content Content Over Years")
plt.xlabel("Year")
plt.ylabel("Number of Titles")
plt.tight_layout()
```

### Analysis:
The graph shows that Netflix had very slow growth before 2015, with minimal content being added. From 2016 onwards, there is a sharp rise, peaking around 2019 with nearly 2000 titles, showing rapid expansion. After that, in 2020–2021, the numbers slightly decline, suggesting Netflix shifted from adding a large volume of content to focusing more on quality.

```python
genre = new_netflix.copy()
genre["Genre"] = genre["Genre"].str.split(", ")
genre = genre.explode("Genre")

# Top 5 genres
top_genres = genre["Genre"].value_counts().head(5).index
top_data = genre[genre["Genre"].isin(top_genres)].reset_index(drop=True)

# Bottom 5 genres
bottom_genres = genre["Genre"].value_counts().sort_values().head(5).index
bottom_data = genre[genre["Genre"].isin(bottom_genres)].reset_index(drop=True)

# 2x2 layout
plt.figure(figsize=(18,12))


# Plot 1: Top Genres vs Type
plt.subplot(2,2,1)

sns.countplot(
    data=top_data,
    y="Genre",
    hue="Type"
)

plt.title("Top Genres by Content Type")
plt.xlabel("Count")
plt.ylabel("Genre")

# Plot 2: Top Genres Over Years
plt.subplot(2,2,2)

year_genre = (
    top_data.groupby(["Year_Added", "Genre"])
    .size()
    .reset_index(name="count")
)

sns.lineplot(
    data=year_genre,
    x="Year_Added",
    y="count",
    hue="Genre",
    marker="o"
)

plt.title("Top Genres Trend Over Years")
plt.xlabel("Year")
plt.ylabel("Number of Titles")


# Plot 3: Bottom Genres
plt.subplot(2,2,3)

sns.countplot(
    data=bottom_data,
    y="Genre",
    color="#A3E4D7"
)

plt.title("Least Popular Genres")
plt.xlabel("Count")
plt.ylabel("Genre")


#  Plot 4: Top Genres Overall 
plt.subplot(2,2,4)

top_counts = genre["Genre"].value_counts().head(5)

sns.barplot(
    x=top_counts.values,
    y=top_counts.index,
    hue=top_counts.index,   
    legend=False
)

plt.title("Top Genres Overall")
plt.xlabel("Count")
plt.ylabel("Genre")
plt.tight_layout()


plt.suptitle("Genre Analysis Across Multiple Dimensions", fontsize=20)
plt.tight_layout()
```

### Analysis:
Netflix has expanded its content over the years, especially after 2016, with a strong focus on popular genres like Drama, Comedy, and International content. Movies dominate most genres, indicating a higher volume of movie production compared to TV shows. While mainstream genres lead in numbers, a few niche categories remain less explored, reflecting targeted audience preferences.

📈 Rapid content growth observed between 2016–2019
🎬 Movies dominate across most genres
🌍 International content is among the most popular
📉 Some genres remain niche eg - Classic & Cult TV
